```python
#Python数据分析（Pandas）
#任务1
import pandas as pd
with open('driver.csv') as f:
    drivers = pd.read_csv(f)
average_driver_rating = drivers['rating'].mean()
print(average_driver_rating)
```

```python
#任务2
bool = drivers['second_language'].notna()
percentage_drivers_with_second_language = bool.sum() / len(drivers)
print(percentage_drivers_with_second_language)
```

```python
#任务3
import numpy as np
l = 0
success = 0
for i in range(1, 5):
    with open(f'rides_{i}.csv', encoding='utf-8') as f:
        rides = pd.read_csv(f)
        l += rides.shape[0]
        success += np.sum(rides['status'] == '成功')
ride_success_rate = success / l
print(ride_success_rate)
```

```python
with open('analysis_results.csv', 'w') as f:
    f.write('insight_type,value\n')
    f.write(f'average_driver_rating,{average_driver_rating:.2f}\n')
    f.write(f'percentage_drivers_with_second_language,{percentage_drivers_with_second_language:.2f}\n')
    f.write(f'ride_success_rate,{ride_success_rate:.2f}\n')
```

```python
#深度学习训练相关名词解释
#1.对每个批次的数据进行归一化处理，将输入的数据的均值调整为0，方差调整为1的标准正态分布，可以加速训练过程，减少过拟合
#2.Dropout是一种防止过拟合的方法，会随机抛弃部分神经元的输出，可以避免神经元之间的过度依赖
#3.梯度裁剪是一种防止梯度消失的方法，当梯度过大的时候，将梯度缩小。可以保证训练的稳定
#4.早停是一种根据验证集性能来判断何时停止训练的方法，当验证集性能不再增加时，会让训练提前结束。因为会在最佳性能的时候停止训练，所以可以防止过拟合
#5.数据增强可以增加数据多样性，例如可以对图像数据进行旋转、翻转、色彩增强
#6.梯度消失是指梯度在反向传播过程中不断缩小，导致训练速度变慢。出现原因是网络层数深，梯度会不断缩小，导致训练速度变慢。解决方法是梯度裁剪、正则化、使用其他激活函数、批量归一化
#7.学习率调度是使用动态的学习率的策略。训练刚开始使用较大学习率，随着训练的进行，学习率逐渐减小，可以提高训练速度和性能
```

```python
#深度学习预测序列
# 假设我们用 60 天的价格预测第 61 天价格
time_steps = 60
features = 1
samples = 200

# 构造输入数据 (samples, time_steps, features)
x_train = np.random.rand(samples, time_steps, features)  # 200个 60天股票价格
y_train = np.random.rand(samples, 1)                     # 200个 对应的第61天股票价格
x_test  = np.random.rand(10, time_steps, features)  
```

```python
# 构建5层Sequential神经网络预测股票价格，要求包含：
# ●2个LSTM层（首层units=4需设置input_shape）
# ●每个LSTM层后接Dropout层(rate=0.2)
# ●最终Dense层(units=1) 实现步骤：
# 1.使用stock_model变量构建模型
# 2.编译模型(loss=‘mean_squared_error’, optimizer=‘adam’)
# 3.训练配置(epochs=5, batch_size=16, verbose=0)
# 4.输出模型summary()
# 5.预测x_test并输出最后一列
import tensorflow as tf
stock_model = tf.keras.Sequential([
    tf.keras.layers.LSTM(units=4, input_shape=(60, 1), return_sequences=True),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.LSTM(4),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1)
])
```

```python
stock_model.compile(loss='mean_squared_error', optimizer='adam')
stock_model.fit(x_train, y_train, epochs=5, batch_size=16, verbose=0)
stock_model.summary()
```

```python
stock_model.predict(x_test)
```
